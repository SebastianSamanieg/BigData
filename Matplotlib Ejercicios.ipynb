## Ejercicio 1

**Siga estos pasos:**
* **Cree un objeto figura llamado fig usando plt.figure()**
* **Use add_axes para agregar un eje al canvas de la figura en [0,0,1,1]. Llame a este nuevo eje ax.**
* **Grafique "plot(x, y)" en ese ejes y configure las etiquetas y títulos para que coincidan con la figura que se presenta a continuación:**

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO
# Datos
x = np.arange(0, 100)
y = x * 2

# Crear un objeto figura llamado fig
fig = plt.figure()

# Agregar un eje al canvas de la figura en [0,0,1,1]
ax = fig.add_axes([0, 0, 1, 1])

# Graficar los datos en el eje 'ax'
ax.plot(x, y, label='y = 2x', color='blue', linestyle='-')

# Configurar etiquetas y títulos
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Gráfico de y = 2x')

# Agregar una leyenda al gráfico
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```

## Ejercicio 4

**Use plt.subplots (nrows=1, ncols=2) para crear el diagrama que se muestra a continuación.**

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO

# Crear un diagrama con dos subgráficos en una fila (1 fila, 2 columnas)
fig, axes = plt.subplots(nrows=1, ncols=2)

# Configurar límites para ambos subgráficos
for ax in axes:
    ax.set_xlim(0.0, 1)
    ax.set_ylim(0.0, 1)

# Graficar en el primer subgráfico
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].set_title('Gráfico 1')

# Graficar en el segundo subgráfico
axes[1].set_xlabel('Eje X')
axes[1].set_ylabel('Eje Y')
axes[1].set_title('Gráfico 2')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar el diagrama con los subgráficos
plt.show()
```

**Ahora grafique (x,y) y (x,z) en esos ejes. Configure adecuadamente los parámétros color, linewidth y linestyle para obtener lo siguiente:**

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO

# Datos
x = np.arange(0,100)
y = x*2
z = x**2

# Crear un diagrama con dos subgráficos en una fila (1 fila, 2 columnas)
fig, axes = plt.subplots(nrows=1, ncols=2)

# Configurar límites para ambos subgráficos
axes[0].set_xlim(0, 100)  # Límites en el primer subgráfico
axes[0].set_ylim(0, 200)
axes[1].set_xlim(0, 100)  # Límites en el segundo subgráfico
axes[1].set_ylim(0, 10000)

# Graficar en el primer subgráfico (x vs. y) - Línea azul punteada
axes[0].plot(x, y, color='blue', linestyle='--', linewidth=2)
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].set_title('Gráfico x vs. y')

# Graficar en el segundo subgráfico solo la línea x^2 en rojo
axes[1].plot(x, z, color='red', linestyle='-', linewidth=2)
axes[1].set_xlabel('Eje X')
axes[1].set_ylabel('Eje Y')
axes[1].set_title('Gráfico x vs. z')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar el diagrama con los subgráficos
plt.show()
```

**Redimensione la figura incluyendo el parámetro figsize al momento de crear la figura en plt.subplots(). Copie y pege el código de la celda anterior para ahorrar tiempo.**

```python
# ESCRIBA SU CODIGO AQUI PARA NO BORRAR EL RESULTADO ESPERADO

# Datos
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Crear un diagrama con dos subgráficos en una fila (1 fila, 2 columnas) y ajustar el tamaño de la figura
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))  # Aquí, figsize ajusta el tamaño de la figura en pulgadas (ancho x alto)

# Configurar límites para ambos subgráficos
axes[0].set_xlim(0, 100)  # Límites en el primer subgráfico
axes[0].set_ylim(0, 200)
axes[1].set_xlim(0, 100)  # Límites en el segundo subgráfico
axes[1].set_ylim(0, 10000)

# Graficar en el primer subgráfico (x vs. y) - Línea azul punteada
axes[0].plot(x, y, color='blue', linestyle='--', linewidth=2)
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].set_title('Gráfico x vs. y')

# Graficar en el segundo subgráfico solo la línea x^2 en rojo
axes[1].plot(x, z, color='red', linestyle='-', linewidth=2)
axes[1].set_xlabel('Eje X')
axes[1].set_ylabel('Eje Y')
axes[1].set_title('Gráfico x vs. z')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar el diagrama con los subgráficos
plt.show()
```
